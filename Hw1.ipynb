```python
import numpy as np
```

```python
print("--- Завдання 1: Одновимірний масив з першими 10 натуральними числами ---")
vector_natural = np.arange(1, 11)
print(f"Вектор з першими 10 натуральними числами:\n{vector_natural}\n")
```

```python
print("--- Завдання 2: Двовимірний масив 3x3, заповнений нулями ---")
matrix_zeros = np.zeros((3, 3))
print(f"Матриця 3x3, заповнена нулями:\n{matrix_zeros}\n")
```

```python
print("--- Завдання 3: Масив 5x5, випадкові цілі числа від 1 до 10 ---")
matrix_rand_int = np.random.randint(1, 11, size=(5, 5))
print(f"Масив 5x5 з випадковими цілими числами від 1 до 10:\n{matrix_rand_int}\n")
```

```python
print("--- Завдання 4: Масив 4x4, випадкові дійсні числа від 0 до 1 ---")
matrix_rand_float = np.random.rand(4, 4)
print(f"Масив 4x4 з випадковими дійсними числами від 0 до 1:\n{matrix_rand_float}\n")
```

```python
print("--- Завдання 5: Два одновимірних масиви 5, випадкові цілі від 1 до 10, поелементні операції ---")
vector_a = np.random.randint(1, 11, size=5)
vector_b = np.random.randint(1, 11, size=5)
print(f"Перший вектор (vector_a):\n{vector_a}")
print(f"Другий вектор (vector_b):\n{vector_b}")
print(f"Додавання (vector_a + vector_b):\n{vector_a + vector_b}")
print(f"Віднімання (vector_a - vector_b):\n{vector_a - vector_b}")
print(f"Множення (vector_a * vector_b):\n{vector_a * vector_b}\n")
```

```python
print("--- Завдання 6: Два вектори розміром 7, скалярний добуток ---")
vector_c = np.random.rand(7) 
vector_d = np.random.rand(7)
print(f"Перший вектор (vector_c):\n{vector_c}")
print(f"Другий вектор (vector_d):\n{vector_d}")
print(f"Скалярний добуток (np.dot(vector_c, vector_d)): {np.dot(vector_c, vector_d)}\n")
```

```python
print("--- Завдання 7: Дві матриці 2x2 та 2x3, перемножити ---")
matrix1 = np.random.randint(1, 11, size=(2, 2))
matrix2 = np.random.randint(1, 11, size=(2, 3))
print(f"Перша матриця 2x2:\n{matrix1}")
print(f"Друга матриця 2x3:\n{matrix2}")
# Матричне множення: кількість стовпців першої має дорівнювати кількості рядків другої
# matrix1 (2x2) @ matrix2 (2x3) -> результат (2x3)
product_matrix = np.dot(matrix1, matrix2)
print(f"Добуток матриць:\n{product_matrix}\n")
```

```python
print("--- Завдання 8: Матриця 3x3, обернена матриця ---")
# Для оберненої матриці потрібно, щоб вона не була сингулярною.
# Генеруємо матрицю, доки вона не буде оборотною (детермінант не дорівнює 0).
while True:
    matrix_inv = np.random.randint(1, 11, size=(3, 3))
    if np.linalg.det(matrix_inv) != 0:
        break
print(f"Початкова матриця:\n{matrix_inv}")
inverse_matrix = np.linalg.inv(matrix_inv)
print(f"Обернена матриця:\n{inverse_matrix}\n")
```

```python
print("--- Завдання 9: Матриця 4x4, транспонувати ---")
matrix_transpose = np.random.rand(4, 4)
print(f"Початкова матриця:\n{matrix_transpose}")
transposed_matrix = matrix_transpose.T
print(f"Транспонована матриця:\n{transposed_matrix}\n")
```

```python
print("--- Завдання 10: Матриця 3x4 та вектор 4, перемножити матрицю на вектор ---")
matrix_mv = np.random.randint(1, 11, size=(3, 4))
vector_mv = np.random.randint(1, 11, size=4)
print(f"Матриця 3x4:\n{matrix_mv}")
print(f"Вектор розміром 4:\n{vector_mv}")
product_mv = np.dot(matrix_mv, vector_mv)
print(f"Добуток матриці на вектор:\n{product_mv}\n")
```

```python
print("--- Завдання 11: Матриця 2x3 та вектор 3, перемножити матрицю на вектор (дійсні числа) ---")
matrix_mv_float = np.random.rand(2, 3)
vector_mv_float = np.random.rand(3)
print(f"Матриця 2x3:\n{matrix_mv_float}")
print(f"Вектор розміром 3:\n{vector_mv_float}")
# Матриця (2x3) @ Вектор (3,) -> результат (2,)
product_mv_float = np.dot(matrix_mv_float, vector_mv_float)
print(f"Добуток матриці на вектор (дійсні числа):\n{product_mv_float}\n")
```

```python
print("--- Завдання 12: Дві матриці 2x2, поелементне множення ---")
matrix_elem1 = np.random.randint(1, 11, size=(2, 2))
matrix_elem2 = np.random.randint(1, 11, size=(2, 2))
print(f"Перша матриця:\n{matrix_elem1}")
print(f"Друга матриця:\n{matrix_elem2}")
elementwise_product = matrix_elem1 * matrix_elem2
print(f"Поелементне множення матриць:\n{elementwise_product}\n")
```

```python
print("--- Завдання 13: Дві матриці 2x2, добуток (матричне множення) ---")
matrix_prod1 = np.random.randint(1, 11, size=(2, 2))
matrix_prod2 = np.random.randint(1, 11, size=(2, 2))
print(f"Перша матриця:\n{matrix_prod1}")
print(f"Друга матриця:\n{matrix_prod2}")
matrix_product = np.dot(matrix_prod1, matrix_prod2) # або matrix_prod1 @ matrix_prod2
print(f"Добуток матриць (матричне множення):\n{matrix_product}\n")
```

```python
print("--- Завдання 14: Матриця 5x5, сума елементів матриці ---")
matrix_sum_elements = np.random.randint(1, 101, size=(5, 5))
print(f"Матриця 5x5:\n{matrix_sum_elements}")
sum_of_elements = np.sum(matrix_sum_elements)
print(f"Сума всіх елементів матриці: {sum_of_elements}\n")
```

```python
print("--- Завдання 15: Дві матриці 4x4, різниця ---")
matrix_diff1 = np.random.randint(1, 11, size=(4, 4))
matrix_diff2 = np.random.randint(1, 11, size=(4, 4))
print(f"Перша матриця:\n{matrix_diff1}")
print(f"Друга матриця:\n{matrix_diff2}")
difference_matrix = matrix_diff1 - matrix_diff2
print(f"Різниця матриць:\n{difference_matrix}\n")
```

```python
print("--- Завдання 16: Матриця 3x3, вектор-стовпчик з сумою елементів кожного рядка ---")
matrix_row_sum = np.random.rand(3, 3)
print(f"Початкова матриця:\n{matrix_row_sum}")
# np.sum(axis=1) обчислює суму по рядках
row_sums_vector = np.sum(matrix_row_sum, axis=1)
# Перетворюємо результат на вектор-стовпчик (розмірність (3,) на (3,1))
row_sums_column_vector = row_sums_vector.reshape(-1, 1)
print(f"Вектор-стовпчик з сумами елементів кожного рядка:\n{row_sums_column_vector}\n")
```

```python
print("--- Завдання 17: Матриця 3x4, матриця з квадратами чисел ---")
matrix_original = np.random.randint(1, 11, size=(3, 4))
print(f"Початкова матриця:\n{matrix_original}")
matrix_squares = matrix_original ** 2 # Поелементне піднесення до квадрату
print(f"Матриця з квадратами чисел:\n{matrix_squares}\n")
```

```python
print("--- Завдання 18: Вектор розміром 4, вектор з квадратними коренями ---")
vector_sqrt_original = np.random.randint(1, 51, size=4)
print(f"Початковий вектор:\n{vector_sqrt_original}")
vector_square_roots = np.sqrt(vector_sqrt_original)
print(f"Вектор з квадратними коренями:\n{vector_square_roots}\n")
```
